```python
import requests
import pandas as pd
import time
from datetime import date

TOKEN = "AtXoVlGaPwlQNAuDoMGGnaNbZRVYVxPB"
headers = {"token": TOKEN}
BASE_URL = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"

stationid = "GHCND:AYM00088963"   # Base Esperanza station
datatypeid = "TAVG"               # average temperature


def fetch_year(year):
    rows = []
    offset = 1
    while True:
        params = {
            "datasetid": "GHCND",
            "stationid": stationid,
            "datatypeid": datatypeid,
            "startdate": f"{year}-01-01",
            "enddate": f"{year+1}-01-01",
            "limit": 1000,
            "offset": offset
        }
        # retry loop for temporary API failures
        for attempt in range(3):
            response = requests.get(BASE_URL, headers=headers, params=params)
            if response.status_code == 200:
                break
            if response.status_code in [429, 503]:
                print(f"[{year}] temporary error {response.status_code}, retrying...")
                time.sleep(2)

        # if response still failed
        if response.status_code != 200:
            print(f"[{year}] failed request:", response.text[:200])
            return pd.DataFrame()

        data = response.json()
        batch = data.get("results", [])

        if not batch:
            break
        rows.extend(batch)
        offset += 1000
        # delay to respect API rate limit
        time.sleep(0.3)
    return pd.DataFrame(rows)

start_year = 1979
end_year = date.today().year

dfs = []

for year in range(start_year, end_year + 1):
    df_year = fetch_year(year)
    if not df_year.empty:
        dfs.append(df_year)
    print(f"Done {year}: {len(df_year)} rows")
    time.sleep(0.3)

df = pd.concat(dfs, ignore_index=True)

# clean dataset
df["date"] = pd.to_datetime(df["date"])
df["temp_c"] = df["value"] / 10

df = df.sort_values("date").reset_index(drop=True)

df.head()
```

```python
"""
NOAA GHCND Antarctic Temperature Data — Cleaning, Preprocessing & EDA
Variable: TAVG (Daily Average Temperature)
Source: NOAA CDO API -> saved CSV (antarctica_temperature_1979_present.csv)

Assumes your CSV has columns like:
date, datatype, station, attributes, value, temp_c
"""

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
import os


OUT = "."
os.makedirs(OUT, exist_ok=True)

sns.set_theme(style="whitegrid", font_scale=1.1)

# -------------------------
# 1) DATA COLLECTION
# -------------------------
print("── Step 1: Data Collection ──")

raw = pd.read_csv(file_path)
print(f"  Raw shape  : {raw.shape}")
print(f"  Columns    : {raw.columns.tolist()}")

# -------------------------
# 2) BEFORE SNAPSHOT
# -------------------------
print("\n── BEFORE snapshot ──")
print(f"  Rows          : {len(raw):,}")
print(f"  Missing values:\n{raw.isnull().sum().to_string()}")
print(raw.head(5).to_string())

# -------------------------
# 3) CLEANING / PREPROCESSING
# -------------------------
print("\n── Step 2: Cleaning ──")

# 1. Drop unnecessary columns: datatype, station, attributes, value
df_dropped = df.drop(['datatype', 'station', 'attributes', 'value'], axis=1)
df_dropped.head()

# 2. z score transformations (mean = 0, SD = 1)
df_dropped["temp_zscore"] = StandardScaler().fit_transform(df[["temp_c"]])
df_dropped.head()


# 3. Find temperature anomaly

df_dropped["month"] = df_dropped["date"].dt.month #Get the month variable
#calculate the temp anomaly - How much warmer or colder a temperature is compared to the typical temperature for that time of year.
df_dropped["temp_anomaly"] = df_dropped["temp_c"] - df_dropped.groupby(["month"])["temp_c"].transform("mean")
df_cleaned = df_dropped.copy()
#drop the month
df_cleaned = df_cleaned.drop(['month'], axis=1)
df_cleaned.head()




```

```python
# -------------------------
# 4) AFTER SNAPSHOT
# -------------------------
print("\n── AFTER snapshot ──")
print(df_cleaned[["date", "temp_c", "temp_zscore", "temp_anomaly"]].head(8).to_string())
```

```python
# -------------------------
# 5) STATISTICAL SUMMARY
# -------------------------
print("\n── Step 3: Statistical Summary ──")

df_cleaned = df_dropped.copy()

summary = df_cleaned.groupby("month")["temp_c"].agg([
    "count", "mean", "median",
    ("std", "std"),
    ("var", "var"),
    "min", "max"
]).round(3)

print(summary.to_string())
summary.to_csv(f"{OUT}/temp_summary_statistics.csv")
```

```python
# -------------------------
# 6) VISUALIZATIONS
# -------------------------
import matplotlib.pyplot as plt
print("\n── Step 4: Visualisations ──")

monthly = df_cleaned.groupby("month")["temp_c"].mean()

fig, ax = plt.subplots(figsize=(13,4.5))

ax.plot(monthly.index, monthly.values, marker="o")

ax.set_xlabel("Month")
ax.set_ylabel("Average Temp (°C)")
ax.set_title("Average Monthly Temperature Pattern")

plt.tight_layout()
plt.savefig(f"{OUT}/fig01_monthly_timeseries.png", dpi=150); plt.close()
print("  ✓ Fig 1 – Monthly time series")
```

```python
# Fig 2: QQ plot (normality)
fig, ax = plt.subplots(figsize=(6, 6))
stats.probplot(df_cleaned["temp_c"].dropna(), dist="norm", plot=ax)
ax.set_title("Fig 2 — QQ Plot (Normality Check)")
plt.tight_layout()
plt.savefig(f"{OUT}/fig02_qqplot.png", dpi=150); plt.close()
print("  ✓ Fig 2 – QQ plot")
```

```python
# Fig 3: Temperature anomalies over time

monthly_anomaly = (
    df_cleaned
    .set_index("date")
    .resample("M")["temp_anomaly"]
    .mean()
)

fig, ax = plt.subplots(figsize=(13,4.5))

ax.plot(monthly_anomaly.index, monthly_anomaly.values, linewidth=1.5)

ax.axhline(0, linestyle="--", linewidth=1)

ax.set_xlabel("Year")
ax.set_ylabel("Temperature Anomaly (°C)")
ax.set_title("Monthly Temperature Anomaly (1979–Present)")

plt.tight_layout()
plt.savefig(f"{OUT}/fig3_anomalies.png", dpi=150); plt.close()
print("  ✓ Fig 3 – Anomalies")
```
